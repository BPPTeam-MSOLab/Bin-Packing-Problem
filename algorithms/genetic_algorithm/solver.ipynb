```python
from optimizer import *
from problem import *
```

```python
def solve(path, seed=10, n_individuals=100, n_elites=10, n_generations=11, p_crossover=0.5, p_mutation=0.3, **kwargs):
	np.random.seed(seed)
	problem = Problem(path)
	objective_function = partial(evaluate, problem=problem)
	Configuration(objective_function, 
			      problem.total_items, 
				  n_individuals=n_individuals, 
				  n_elites=n_elites, 
				  n_generations=n_generations, 
				  p_crossover=p_crossover, 
				  p_mutation=p_mutation, 
				  problem=problem, 
				  verbose=True)
	frequency = kwargs.get('frequency', 10)
	use_tqdm = kwargs.get('use_tqdm', True)
	Optimizer(frequency=frequency, use_tqdm=use_tqdm).optimize()
	problem.visualize()
	Configuration().reset()
```

```python
solve('../../Data/Dataset/20_1_1.dat', 0)
```
